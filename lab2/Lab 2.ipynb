```python
#list
grades = [85, 90, 78, 92, 88, 94,22.5,87.6]
print("list of grades:", grades)
print(type(grades))
print(grades[1])
print(grades[0])
print(grades[2])
#change value
grades[0] = 58
print("list of grades:", grades)
#length 
print(len(grades))
#negative indexing
print(grades[-2])
print(grades[-3])
```

```python
#tuple
student = ("Ali", 20, "Data Science")
print("student tuple:", student)
print(type(student))
print(student[0])
print(student[1])
print(student[2])
#lenght
print(len(student))
#negative indexing
print(student[-1])
print(student[-2])
```

```python
#Dictionary
student_grades = {
    "Ali": 85,
    "Sara": 90,
    "Ahmed": 78,
    "Rayan": 59,
    "Rehan": 95
}
print("student grades Dictionary:", student_grades)
print(student_grades ["Ali"])
print(student_grades ["Ahmed"])
print(student_grades.get("Rayan"))
print(type(student_grades))
print(student_grades.keys())
print(student_grades.values())
print(student_grades.items())
    
```

```python
#sets
grades_set = {85, 90, 78, 92, 88, 90, 94, 22.5, 87.6, 22.5}
print(grades_set)
print(type(grades_set))
print(len(grades_set))
for values in grades_set:
    print(values)
```
