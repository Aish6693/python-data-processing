```python
# write students grade in text file
with open("students.txt", "w") as file:
    file.write("Ali 85\n")
    file.write("Sara 90\n")
    file.write("Ahmed 78\n")
    file.write("Reeha 95\n")
    file.write("Rehan 75\n")

print("Student grades save in text file")

# read text file
with open("students.txt", "r") as file:
    data = file.read()

print("Student Grades:")
print(data)
```

```python
# Write CSV file
import csv

data = [
    ["Name", "Grade"],
    ["Ali", 85],
    ["Sara", 90],
    ["Ahmed", 78],
    ["Reeha", 95],
    ["Rehan", 75]
]

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print("CSV file is created")

#Read CSV file
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
#Write JSON file
import json

students = {
    "Ali": 85,
    "Sara": 90,
    "Ahmed": 78,
    "Reeha": 95,
    "Rehan": 75
}

with open("students.json", "w") as file:
    json.dump(students, file)

print("JSON file is created")

#Read JSON file
import json

with open("students.json", "r") as file:
    data = json.load(file)

print("Student Grades:")
print(data)
```

```python
#List Comprehension
grades = [45, 70, 82, 30, 60]

# 5 marks bonus
bonus_grades = [g + 5 for g in grades]

# Pass/Fail result
results = ["Pass" if g >= 50 else "Fail" for g in grades]

print("Original Grades:", grades)
print("Bonus Grades:", bonus_grades)
print("Results:", results)
```

```python
#Error handling
import csv

try:
    with open("students.csv","r") as file:
        reader = csv.reader(file)

        next(reader)  

        for row in reader:
            name = row[0]
            grade = int(row[1])   
            
            print("Student:", name)
            print("Grade:", grade)
            print("----------------")

except FileNotFoundError:
    print("Error: CSV file not found")

except ValueError:
    print("Error: No Grade number")

finally:
    print("Program finish successfully")
```
