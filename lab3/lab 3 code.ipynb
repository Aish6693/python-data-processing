```python
import pandas as pd

file_path = r"C:\Users\HP\.cache\kagglehub\datasets\jainilcoder\netflix-stock-price-prediction\versions\1\NFLX.csv"

df = pd.read_csv(file_path)

print(df.head())
```

```python
import numpy as np

# Close price column ko NumPy array mein convert
prices = df['Close'].values

# Mean
print("Mean:", np.mean(prices))

# Std
print("Std:", np.std(prices))

# Daily change
change = np.diff(prices)
print("Change:", change[:5])
```

```python
#array
arr1 = np.array(prices[:5])
arr2 = arr1.reshape(5,1)
arr3 = arr2.reshape(5,1,1)

print("\n1D Array:", arr1)
print("\n2D Array:\n", arr2)
print("\n3D Array:\n", arr3)
```

```python
#Indexing & Slicing
print("\nFirst element:", arr1[0])
print("Last 3 elements:", arr1[-3:])
print("All rows second column (2D):", arr2[:,0])
```

```python
#reshaping
reshaped = prices[:10].reshape(5,2)
print("\nReshaped (5x2):\n", reshaped)
```

```python
#Mathematical Operations
print("\nAdd 10:", arr1 + 10)
print("Multiply by 2:", arr1 * 2)
```

```python
#Statistical Functions
print("\nMean:", np.mean(prices))
print("Median:", np.median(prices))
print("Std Dev:", np.std(prices))
print("50th Percentile:", np.percentile(prices, 50))
```

```python
#Linear Algebra
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("\nDot Product:\n", np.dot(A,B))
print("Matrix Multiplication:\n", A @ B)
```

```python
#Random Numbers
print("\nRandom Numbers:", np.random.rand(3))
print("Random Integers:", np.random.randint(1,10,5))
```

```python
#Stock Analysis
change = np.diff(prices)

print("\nDaily Changes (first 5):", change[:5])
print("Max Price:", np.max(prices))
print("Min Price:", np.min(prices))
```
